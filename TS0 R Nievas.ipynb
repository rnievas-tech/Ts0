## Introduccion teorica
### Señal sinusoidal
Una señal sinusoidal es una forma de onda periódica que puede representarse mediante la funcion seno o coseno.Su expresion general en un dominio continuo es:
\begin{equation}
x(t)=Vm+A*sen(2*\pi*f*t+φ)
\end{equation}
donde

A[V]:Amplitud de la señal,indica el valor máximo de la señal respecto al eje horizontal 

Vm[V]:Valor medio de la señal,corresponde al promedio de todos los valores en intervalo 

f[Hz]:Frecuencia,indica cuantos ciclos completos ocurren por segundo

t[s]:Variación temporal 

φ[rad]:Fase que representa un corrimiento en el tiempo de la señal 

### Señales en tiempo discreto que puedan 
En el análisis de señales, es necesario convertir señales analógicas, definidas en tiempo continuo $x(t)$, en señales discretas que puedan ser procesadas digitalmente.

Este proceso se denomina muestreo y consiste en tomar valores de la señal en instantes separados por un período $T$.
Para generar una señal discreta se procede a muestrear una señal continua en tiempo continuo con periodo T,esto consiste en multiplicar esta señal por un tren de impulsos donde cada impulso toma como amplitud el valor de señal original en los instantes de muestreo
\begin{equation}
xs(t)=x(t)*s(t)
\end{equation}

### Frecuencia en tiempo discreto
El muestreo produce una repetición periódica del espectro de la señal, separada por la frecuencia de muestreo $f_s$. Si la frecuencia de muestreo es demasiado baja, las réplicas del espectro se superponen lo que impide distinguir las componentes originales de las réplicas.
Para evitarlo, se debe cumplir el criterio de Nyquist:

\begin{equation}
fs>2fb
\end{equation}

Cuando la frecuencia de la señal se aproxima a $fb$, la señal muestreada comienza a presentar menos puntos por período, lo que dificulta su representación.

Si la frecuencia supera este valor, donde la señal observada en tiempo discreto aparenta tener una frecuencia diferente a la original.


## Analisis y desarrollo
En el siguiente trabajo se empieza a utilizar el leguaje python generando una funcion seno con parametros previamente definidos, considerando el muestreo de la señal
Como primer paso se prepara el entorno que se desea trabajar importando las librerias necesaria y definiendo las funciones para lograr



```python
import matplotlib.pyplot as plt
import numpy as np
from numpy.fft import fft
import scipy.signal as sp
import scipy.signal.windows as windows

def eje_temporal (N, fs):
    tt = np.arange(N) / fs
    return tt


def func_senoidal (tt, frec, amp, fase , v_medio ):
    xx = amp * np.sin (2 * np.pi * frec * tt + fase) + v_medio 
    return xx

amp = 1 # amplitud en V
N = 1000 #Cantidad de muestra
fs = 1000
fase=0
v_medio=0
frec = 1


tt = eje_temporal (N = N, fs = fs)
s1 = func_senoidal (tt , frec, amp ,fase,v_medio)

######################GRAFICO######################################


plt.plot(tt, s1, color='orange',label='Señal senoidal')
plt.xlabel("Tiempo [seg]")
plt.ylabel("Y(t) [V]")
plt.title("Figura 1:Senoidal en funcion del tiempo")
plt.xlim([0,max(tt)])
plt.grid(True)
plt.legend()
plt.show()
```
